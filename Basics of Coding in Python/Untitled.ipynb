## Getting Data into Python

```python
#  Getting Data into Python

# Now we will discuss how to get a data set into Python.

# We will use the read_csv command, using the following structure:
# your_name_for_the_data_set = pd.read_csv("the_path_to_the_file")

# The command read_csv comes from the pandas package. Following the code

import pandas as pd
# After importing the pandas package, we can access the read_csv command by typing pd.
# To use the read_ csv() command, type pd followed by a period, then the command read_csv.
#The next part of the code is the path to the data file, contained in double quotes.
# For example, We need to import the california_housing_training data set. The code to import the data set is given below.
house_train = pd.read_csv("/content/sample_data/california_housing_train.csv")
```

## Saving Output R/Python

```python
# Saving Output R/Python

#w let
# Let us illustrate the difference between running a command and not saving the output
# versus running a command and saving the output, using a contingency table.
# If we make a contingency table and save it in order to make a bar chart using that table.
# The code to make a contingency table is one line of code, given below

pd.crosstab(house_train['population'], house_train['households'])

# If you do not save the output generated by the crosstab() command, the resulting contingency table is displayed.
# If we only wanted to make a table, and not use it elsewhere in our code, this would be sufficient.
# However, we want to use this table to make a bar chart.
# Therefore, we have to save it. To save it, we add a name for the saved item and an
# equal sign to the left of the command that generated the output.
# The code is given below
crosstab_1  = pd.crosstab(house_train['population'], house_train['households'])
```

## Accessing Records and Variables in Python

### 1.   Accessing single record

```python
#  Accessing Records and Variables in Python

# We use the loc attribute, which all pandas data frames have,
# and state what part of the data frame you want to see.
# Python references its records starting at record zero, so if we want to view the first record we request record 0.
# Similarly, to view the second record, request record 1, and so on.
# For example, to see the first record of bank_train, use the following code:

house_train.loc[0]

#Using the .loc attribute as above to view the first record will return the values of all variables for that record.
# Output shows the first four variable values for the first record in the california_housing_train data set.
```


### 2.  Accessing more than one record

```python
#  Accessing Records and Variables in Python

# What if we want to access many records?
# We will use the .loc attribute and list the records we want to see.
# If you want the first, third, and fourth rows, use the following code:
house_train.loc[[0, 2, 3]]
```

### 3. Accessing first N records

```python
#  Accessing Records and Variables in Python

# We want the first 6 rows, inclusive, we would use the following code:
house_train[0:6]
```

### 4. Accessing records by specifying variables

```python
#  Accessing Records and Variables in Python

# While we reference the rows by numbers, our columns have names.
# This means that if we want to specify which variable we want to see, we give its name.

house_train['total_rooms']

# Using one set of brackets and putting the variable name in single quotes returns the entire variable
```

### 5. Accessing records using mujltiple variables

```python
#  Accessing Records and Variables in Python

# What if we want to see multiple variables?
# We will use the .loc attribute and list the variables we want to see.
# If we want to see the age and job variables,
# we put each of those variable names inside single quotes, inside double brackets,
# separated by commas.
house_train[['longitude','latitude']]
```
